```python
# 조합의 수식
# nCr : n! / r!(n-r)!
import copy

candidate_key_combination_list = list()

def create_combination(used_item_list, unused_item_list, now_idx, more_use):
    if more_use == 0:
        global candidate_key_combination_list
        candidate_key_combination_list.append("".join(map(str,used_item_list)))

    for idx in range(now_idx, len(unused_item_list)):
        new_unused_item_list = copy.deepcopy(unused_item_list)
        new_used_item_list = copy.deepcopy(used_item_list)
        
        new_used_item_list.append(new_unused_item_list.pop(idx))
        new_more_use = more_use - 1
        
        create_combination(new_used_item_list, new_unused_item_list, idx, new_more_use)
        
def check_minimality(uniqueness_candidate_key):
    minimality_candidate_key = copy.deepcopy(uniqueness_candidate_key)
    for test_key in uniqueness_candidate_key:
        is_remove = list()
        for repeat_key in minimality_candidate_key:
            temp_list = list()
            for alpha in test_key:
                if alpha in repeat_key:
                    temp_list.append(True)
                else:
                    temp_list.append(False)
            is_remove.append(temp_list)
        
        is_remove = [idx for idx, check_list in enumerate(is_remove) if all(check_list)]
        if len(is_remove) >= 2:
            for remove_idx in reversed(is_remove):
                minimality_candidate_key.pop(remove_idx)
            minimality_candidate_key.append(test_key)
        else:
            pass
    return minimality_candidate_key
    

def solution(relation):
    global candidate_key_combination_list
    candidate_key_hash_dict = dict()

    # 4C1 -> 4
    # 4C2 -> 6
    # 4C3 -> 4
    # 4C4 -> 1    
    create_combination_list = list(range(len(relation[0])))
    for combination_length in range(1, len(relation[0])+1):
        create_combination(list(), create_combination_list, 0, combination_length)

    for combination_element in candidate_key_combination_list:
        candidate_key_hash_dict[combination_element] = set()
        
    for hash_key in candidate_key_hash_dict:
        for item in relation:
            input_data = ""
            for alpha in hash_key:
                input_data += item[int(alpha)]
            candidate_key_hash_dict[hash_key].add(input_data)
    
    uniqueness_candidate_key = list()
    for hash_key in candidate_key_hash_dict:
        if len(candidate_key_hash_dict[hash_key]) == len(relation):
            uniqueness_candidate_key.append(hash_key)
    
    minimality_candidate_key = check_minimality(uniqueness_candidate_key)
    
    answer = len(minimality_candidate_key)
    
    return answer
```
