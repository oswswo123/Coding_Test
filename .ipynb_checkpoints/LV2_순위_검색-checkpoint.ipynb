```python
def search_start_idx(condition_list, condition_score):
    head_idx = 0
    tail_idx = len(condition_list)
    pivot_idx = None
    is_same = False
    
    while head_idx < tail_idx:
        pivot_idx = (head_idx + tail_idx) // 2
        if condition_score <= condition_list[pivot_idx]:
            tail_idx = pivot_idx
        else:
            head_idx = pivot_idx + 1
            
    return tail_idx

def str_to_int(element_str):
    element_str[-1] = int(element_str[-1])
    return element_str

def solution(info, query):
    language = ["cpp", "java", "python", "-"]
    job_group = ["backend", "frontend", "-"]
    career = ["junior", "senior", "-"]
    soulfood = ["chicken", "pizza", "-"]
    hash_key = list()
    
    for lan_item in language:
        for jg_item in job_group:
            for car_item in career:
                for sf_item in soulfood:
                    hash_key.append(lan_item + jg_item + car_item + sf_item)
    
    hash_dict = {hash_key_element:list() for hash_key_element in hash_key}
    
    data_list = [element_str.split() for element_str in info]
    data_list = list(map(lambda x : str_to_int(x), data_list))
    for data_element in data_list:
        element_hash = list()
        lan, job, car, sf, score = data_element
        for lan_item in language:
            if lan_item == lan or lan_item == "-":
                for jg_item in job_group:
                    if jg_item == job or jg_item == "-":
                        for car_item in career:
                            if car_item == car or car_item == "-":
                                for sf_item in soulfood:
                                    if sf_item == sf or sf_item == "-":
                                        element_hash.append(lan_item + jg_item + car_item + sf_item)
        for each_hash in element_hash:
            hash_dict[each_hash].append(score)
    
    for key in hash_dict:
        hash_dict[key].sort()
    
    query = [element_str.replace(" and ", "") for element_str in query]
    query_list = [element_str.split() for element_str in query]
    
    result = list()
    for query_element in query_list:
        condition_score = int(query_element[-1])
        condition_else = query_element[0]
        
        condition_list = hash_dict[condition_else]
        bn_search_result = search_start_idx(condition_list, condition_score)

        result.append(len(condition_list) - bn_search_result)

    return result
```
