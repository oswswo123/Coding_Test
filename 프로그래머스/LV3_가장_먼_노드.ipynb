```python
# Dijkstra 문제인 줄 앎 (무조건 시간초과)
# 그냥 BFS 문제였음

import heapq
import collections

class Graph:
    def __init__(self, num_vertex, num_edge):
        self.num_vertex = num_vertex
        self.num_edge = num_edge
        self.adjacency_list = [list() for _ in range(num_vertex)]
    
    def add_edge(self, first_vertex, second_vertex):
        self.adjacency_list[first_vertex].append(second_vertex)
        self.adjacency_list[second_vertex].append(first_vertex)
    
    def adjacency_sorting(self):
        for idx in range(self.num_vertex):
            self.adjacency_list[idx] = sorted(self.adjacency_list[idx])
            
    def bfs(self, start_vertex):
        INF = 999999
        queue = collections.deque(list())
        self.bfs_dist = [INF for _ in range(self.num_vertex)]
        self.bfs_dist[start_vertex] = 0
        queue.append(start_vertex)
        
        while bool(queue):
            now_visit = queue.popleft()
            for adjacency_vertex in self.adjacency_list[now_visit]:
                if self.bfs_dist[adjacency_vertex] == INF:
                    queue.append(adjacency_vertex)
                    self.bfs_dist[adjacency_vertex] = self.bfs_dist[now_visit] + 1
    
    def dijkstra(self, start_vertex):
        INF = 999999
        self.distance_list = [INF for _ in range(self.num_vertex)]
        self.previous_vertex = [None for _ in range(self.num_vertex)]
        self.distance_list[start_vertex] = 0
        
        heap = list()
        for idx in range(self.num_vertex):
            heapq.heappush(heap, (self.distance_list[idx], idx))
        
        while bool(heap):
            now_distance, now_vertex = heapq.heappop(heap)
            for adjacency_vertex in self.adjacency_list[now_vertex]:
                if now_distance + 1 < self.distance_list[adjacency_vertex]:
                    self.distance_list[adjacency_vertex] = now_distance + 1
                    self.previous_vertex[adjacency_vertex] = now_vertex
                    change_idx = [idx for idx, element in enumerate(heap) if element[1] == adjacency_vertex]
                    heap[change_idx[0]] = (self.distance_list[adjacency_vertex], adjacency_vertex)
                    heapq.heapify(heap)

def solution(n, edge):
    graph = Graph(n, len(edge))
    
    for first_vertex, second_vertex in edge:
        graph.add_edge(first_vertex-1, second_vertex-1)
    
    # graph.dijkstra(0)  
    # largest_distance = max(graph.distance_list)
    # answer = graph.distance_list.count(largest_distance)
    
    graph.bfs(0)
    largest_distance = max(graph.bfs_dist)
    answer = graph.bfs_dist.count(largest_distance)
    
    return answer
```
