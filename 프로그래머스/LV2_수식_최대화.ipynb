```python
import re
import copy

operator_priority = list()
calculation_result = set()

def create_operator_priority(used_operator, unused_operator, combination_length):
    if combination_length == 3:
        operator_priority.append(used_operator)
        return
    
    for operator in unused_operator:
        new_used_operator = copy.deepcopy(used_operator)
        new_unused_operator = copy.deepcopy(unused_operator)
        
        new_used_operator.append(operator)
        new_unused_operator.remove(operator)
        
        create_operator_priority(new_used_operator, new_unused_operator, len(new_used_operator))
        
def do_calculation(polynomial):
    global operator_priority
    global calculation_result
    
    for now_priority in operator_priority:
        result = list()
        temp_polynomial = copy.deepcopy(polynomial)
        
        for now_priority_operator in now_priority:
            
            while(now_priority_operator in temp_polynomial):
                operator_idx = temp_polynomial.index(now_priority_operator)
                insert_idx = operator_idx - 1
                
                first_operand = temp_polynomial.pop(insert_idx)
                operator = temp_polynomial.pop(insert_idx)
                second_operand = temp_polynomial.pop(insert_idx)
                
                if operator == "+":
                    cal_result = first_operand + second_operand
                elif operator == "-":
                    cal_result = first_operand - second_operand
                elif operator == "*":
                    cal_result = first_operand * second_operand
                else:
                    print("error")
                    
                temp_polynomial.insert(insert_idx, cal_result)

        calculation_result.add(abs(temp_polynomial[0]))

def solution(expression):
    global calculation_result
    exist_operator = "+*-"
    
    list_operator = re.findall("[+\-*]", expression)
    list_operand = re.findall("[0-9]{1,}", expression)
    
    polynomial = [int(list_operand[idx//2]) if idx%2==0 else list_operator[idx//2] for idx in range(len(list_operator) + len(list_operand))]
    
    create_operator_priority(list(), list(exist_operator), 0)
    do_calculation(polynomial)
    
    answer = max(calculation_result)

    return answer
```
