```python
def check_distance(room):
    candidate_pos = list()
    partition_pos = list()
    must_check_pos = list()
    
    for row_idx, row in enumerate(room):
        for col_idx, col in enumerate(row):
            if col == "P":
                candidate_pos.append([row_idx, col_idx])
            elif col == "X":
                partition_pos.append([row_idx, col_idx])
    
    for idx in range(len(candidate_pos)-1):
        for another_idx in range(idx+1, len(candidate_pos)):
            manhattan_distance = abs(candidate_pos[idx][0] - candidate_pos[another_idx][0]) \
                            + abs(candidate_pos[idx][1] - candidate_pos[another_idx][1])
            if manhattan_distance <= 2:
                if manhattan_distance <= 1:
                    # 맨하튼 거리 1 이하면 무조건 거리두기 못지킴
                    return 0
                else:
                    must_check_pos.append([candidate_pos[idx], candidate_pos[another_idx]])

    for check_set in must_check_pos:
        if check_set[0][0] == check_set[1][0]:
            # Y축이 같음
            test_pos = [check_set[0][0], check_set[0][1]+1]
            if test_pos not in partition_pos:
                return 0
        elif check_set[0][1] == check_set[1][1]:
            # X축이 같음
            test_pos = [check_set[0][0]+1, check_set[0][1]]
            if test_pos not in partition_pos:
                return 0
        elif check_set[0][1] < check_set[1][1]:
            # 오른쪽 대각선 방향
            test_pos_down = [check_set[0][0]+1, check_set[0][1]]
            test_pos_right = [check_set[0][0], check_set[0][1]+1]
            if (test_pos_right not in partition_pos) or (test_pos_down not in partition_pos):
                return 0
        else:
            # 왼쪽 대각선 방향
            test_pos_down = [check_set[0][0]+1, check_set[0][1]]
            test_pos_left = [check_set[0][0], check_set[0][1]-1]
            if (test_pos_left not in partition_pos) or (test_pos_down not in partition_pos):
                return 0

    return 1  

def solution(places):
    answer = list()
    
    for each_room in places:
        answer.append(check_distance(each_room))

    return answer
```
